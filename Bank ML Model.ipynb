```python
# Importing key Python Libraries for the Data Analysis and Data Training Model

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score, f1_score
```

```python
# Loading Dataset 

df = pd.read_csv('bank.csv')
```

```python
# Converting Categorical Data into Numbers 

from sklearn.preprocessing import LabelEncoder

code = LabelEncoder()

df['job'] = code.fit_transform(df['job'])
df['marital'] = code.fit_transform(df['marital'])
df['education'] = code.fit_transform(df['education'])
df['month'] = code.fit_transform(df['month'])
df['poutcome'] = code.fit_transform(df['poutcome'])
df['contact'] = code.fit_transform(df['contact'])
df['deposit'] = code.fit_transform(df['deposit'])
df['loan'] = code.fit_transform(df['loan'])
df['housing'] = code.fit_transform(df['housing'])
df['default'] = code.fit_transform(df['default'])
```

```python
X = df.drop(columns=['deposit'])
y = df['deposit']
```

```python
X 
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.2, random_state=0)
```

```python
len(X_train)/len(df)*100
```

```python
model = LogisticRegression()

model
```

```python
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

y_pred
```

```python
print('Accuracy Score of Model:', accuracy_score(y_test, y_pred))
```

```python
print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
```

```python
print('Precision Score: ', precision_score (y_test, y_pred))
```

```python
print('F1 Report: ', f1_score (y_test, y_pred))
```

```python
print('Classification Report: ', classification_report (y_test, y_pred))
```
